```python
import pandas as pd

df = pd.read_csv('C:/Users/alimc/anaconda3/amazon_order_history.csv.')
df.head()
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head()
```

```python
df["total"] = df["total"].str.replace('=SUBTOTAL(109,F2:F5)', '0' ).astype(float)
df.head()
```

```python
df["total"].sum()
```

```python
df["total"].mean()
```

```python
df["total"].median()
```

```python
df["total"].max()
```

```python
df["total"].min()
```

```python
df["tax"] = df["tax"].str.replace('=SUBTOTAL(109,J2:J5)', '0' ).astype(float)
df.head()
```

```python
df["tax"].sum()
```

```python
df.drop(df.tail(1).index,inplace=True)

df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
df.plot.bar(x='date', y='total', rot=90,)
```

```python
daily_orders = df.groupby('date').sum()["total"]
daily_orders.head()
```
